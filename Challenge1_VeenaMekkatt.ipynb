```python
import pandas as pd
data = pd.read_csv('C:/Users/dell/Downloads/company_sales_data.csv')
pd.set_option('display.max_rows', None)
data
```

```python
df=data[data['facewash']<1500]
df
```

```python
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns  
sns.set_style("whitegrid")
X = df['month']
YFC = df['facecream']
ZFW = df['facewash']
  
X_axis = np.arange(len(X))
plt.bar(X_axis + 0.2, YFC, 0.4, label = 'Face Cream Sales',color='#963336',linestyle="dashed")
plt.bar(X_axis - 0.2, ZFW, 0.4, label = 'Face Wash Sales',color='#E3CE57',linestyle="dashed")
  
plt.xticks(X_axis, X)
plt.xlabel("Month")
plt.ylabel("Sales units in number")
plt.suptitle("Facewash Sales Vs Facecream Sales",y=1.05, fontsize=14)
plt.title("For the months facewash sales less than 1500",fontsize=12)
plt.legend()
plt.show()
```
